#TASK 1 and Task 2


```python
#task1
import pandas as pd
import numpy as np

List = ['Monday','Tuesday','Wednesday','Thursday','Friday']

for i in List:
  sheets = pd.read_excel('TimeTable, FSC, Fall-2022.xlsx', sheet_name = None, header = 1, skiprows = [1,2,3])
  sheets[i].dropna(axis = 1, how = 'all', inplace = True)
  x = sheets[i].keys()
  for k in x:
    if len(k) == 10:
      for j in sheets[i][k]:
        if len(str(j)) > 4:
        # y = sheets['Monday'].replace('Unamed: 8','12:30')
          sheets[i][k] = sheets[i][k].replace(j,k[18:])
```

TASK 3


```python
#task 3

check = input("Enter your subject to check it's schedule: ")
for l in List:
  x = sheets[l].keys()
  for i in x:
    for j in range(len(sheets[l][i])):
      if check == sheets[l][i][j]:
        print("Room: ",sheets[l]['Room'][j]," Time of class: ",i,"day: ",l)

```

TASK 4

```python
#task 4
subs = []
num = int(input("Enter how many subjects you wanna know it's schedule: "))

for i in range(num):
  xcheck = input("Enter your subject here: ")
  subs.append(xcheck)

for k in subs:
  for l in List:
    x = sheets[l].keys()
    for i in x:
      for j in range(len(sheets[l][i])):
        if k == sheets[l][i][j]:
          print("Subject: ",k,"Room: ",sheets[l]['Room'][j]," Time of class: ",i,"day: ",l)
```

TASK 5

```python
#task5
print("Empty Slots: ")
for l in List:
  x = sheets[l].keys()
  for i in x:
    for j in range(len(sheets[l][i])):
      if str(sheets[l][i][j]) == "nan":
        print("Room: ",sheets[l]['Room'][j]," Time of class: ",i,"day: ",l)
```

TASK 6

```python
#task 6

room = ""
maximum = 1

for l in List:
  size = len(sheets[l])
  for i in range(size):
    count = sheets[l].loc[[i]].isna().sum().sum()
    if count >= maximum:
      maximum = count
      room = sheets[l]['Room'][i]
print("Classroom which is less busy overrall week is ",room)
```
